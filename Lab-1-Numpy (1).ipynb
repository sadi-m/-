**Задание 1.** Создайте одномерный вектор `a` целых значений от $l_1$ до $l_2$ включительно. На его основе вычислите следующие вектора и матрицы (размер вектора равен $n = l_2 - l_1 + 1$):

* одномерный вектор длины $n$, состоящий из элементов `a`, расположенных в обратном порядке;
* одномерный вектор длины $n$, состоящий из значений 0 и 1, в котором значение 1 соотвествует элементам, делящимся на 3 в исходном векторе `a`;
* двумерную матрицу-строку размера 1 на $n$, содержащую элементы вектора `a`;
* двумерную матрицу-столбец размера $n$ на 1, содержащую элементы вектора `a`;
* двумерную матрицу размера $n$ на $n$, содержащую все попарные произведения элементов вектора `a`;
* двумерную матрицу размера $n+2$ на $n+2$, основанную на матрице из предыдущего пункта и дополненную "бордюром" из нулей, окружающим ненулевые значения;
* двумерную матрицу размера $n$ на $n$, содержащую элементы вектора `a` на побочной диагонали;
* двумерную матрицу размера $n$ на 3, содержащую повторяющиеся значения из вектора `a` в её столбцах;
* двумерную матрицу размера 5 на $n$, содержащую повторяющиеся значения из вектора `a` в её строках.

Каждый из пунктов должен выводить на экран соответствующее значение.

```python
#Этот код создает и обрабатывает вектор a, состоящий из целых чисел от 1 до 10, а затем выполняет несколько операций с ним. Вот краткое описание каждого шага:
#1Создание вектора a: Вектор a содержит числа от 1 до 10.
#2Обратный порядок: Вектор a разворачивается в обратном порядке.
#3Метка деления на 3: Создается вектор из 0 и 1, где 1 обозначает, что соответствующий элемент вектора a делится на 3 без остатка.
#4Матрица-строка: Преобразование вектора в строковую матрицу (размер 1×n).
#5Матрица-столбец: Преобразование вектора в столбцовую матрицу (размер n×1).
#6Матрица попарных произведений: Создается матрица, в которой каждый элемент — это произведение пар чисел из вектора a.
#7Матрица с бордюром из нулей: К матрице попарных произведений добавляется бордюр (рамка) из нулей, увеличивая размер матрицы на 2 по каждой стороне.
#8Матрица с побочной диагональю: Создается матрица, в которой элементы вектора a располагаются на побочной диагонали (слева направо).
#9Матрица с повторяющимися столбцами: Вектор a повторяется три раза в виде столбцов.
#10Матрица с повторяющимися строками: Вектор a повторяется пять раз в виде строк.
import numpy as np

# Задаем значения l1 и l2
l1, l2 = 1, 10  # например, от 1 до 10

# Создаем вектор a
a = np.arange(l1, l2 + 1)

# 1. Вектор, состоящий из элементов a в обратном порядке
a_reversed = a[::-1]

# 2. Вектор, состоящий из значений 0 и 1, где 1 соответствует элементам, делящимся на 3
a_div3 = (a % 3 == 0).astype(int)

# 3. Матрица-строка размера 1 на n
a_row_matrix = a.reshape(1, -1)

# 4. Матрица-столбец размера n на 1
a_col_matrix = a.reshape(-1, 1)

# 5. Матрица размера n на n, содержащая все попарные произведения элементов вектора a
a_product_matrix = a[:, np.newaxis] * a[np.newaxis, :]

# 6. Матрица размера n+2 на n+2, дополненная бордюром из нулей
a_bordered_matrix = np.pad(a_product_matrix, pad_width=1, mode='constant', constant_values=0)

# 7. Матрица размера n на n, содержащая элементы вектора a на побочной диагонали
a_anti_diag_matrix = np.zeros((len(a), len(a)))
np.fill_diagonal(np.fliplr(a_anti_diag_matrix), a)

# 8. Матрица размера n на 3, содержащая повторяющиеся значения из вектора a в её столбцах
a_repeated_cols = np.tile(a, (3, 1)).T

# 9. Матрица размера 5 на n, содержащая повторяющиеся значения из вектора a в её строках
a_repeated_rows = np.tile(a, (5, 1))

# Выводим результаты
print("Вектор a:\n", a)
print("Вектор в обратном порядке:\n", a_reversed)
print("Вектор 0 и 1 (деление на 3):\n", a_div3)
print("Матрица-строка:\n", a_row_matrix)
print("Матрица-столбец:\n", a_col_matrix)
print("Матрица попарных произведений:\n", a_product_matrix)
print("Матрица с бордюром:\n", a_bordered_matrix)
print("Матрица с побочной диагональю:\n", a_anti_diag_matrix)
print("Матрица с повторяющимися столбцами:\n", a_repeated_cols)
print("Матрица с повторяющимися строками:\n", a_repeated_rows)
```

Используйте для этого генератор из модуля `numpy.random`. Сделайте следующее:

* Вычислите двумерную матрицу размера $n$ на $m$, в которой все отрицательные значения заменены на 0;
* Вычислите двумерную матрицу `Y` размера $n$ на $m$, каждый элемент которой вычисляется следующим образом: $y_{ij} = sin(x_{ij}) + cos(x_{ji})$;
* Предложите не менее четырех способов выделения целочисленной части из значений матрицы `X`;
* Вычислите двумерную матрицу размера $n$ на 3, содержащую в первом столбце суммы элементов, во втором столбце средние значения, а в третьем столбце среднеквадратичные отклонения для каждой строки матрицы `X`.

```python
#Этот код создает случайную матрицу X (размером 5x4) с нормально распределенными значениями и выполняет следующие операции:
#Заменяет отрицательные значения в матрице X на 0.
#Вычисляет матрицу Y, используя синус от X и косинус от матрицы с неотрицательными значениями.
#Извлекает целочисленную часть элементов матрицы X разными способами: округление вниз, вверх, усечение и стандартное округление.
#Создает матрицу с суммами, средними значениями и стандартными отклонениями по строкам матрицы X.
#Каждая операция обрабатывает данные матрицы X различными методами.
import numpy as np 

# Задаем размеры матрицы
n, m = 5, 4  # например, 5 на 4

# Создаем матрицу X с нормально распределенными случайными величинами
X = np.random.normal(loc=0, scale=1, size=(n, m))

# 1. Вычисляем матрицу, в которой все отрицательные значения заменены на 0
X_non_negative = np.where(X < 0, 0, X)

# 2. Вычисляем матрицу Y (без транспонирования)
Y = np.sin(X) + np.cos(X_non_negative)

# 3. Способы выделения целочисленной части из значений матрицы X
X_floor = np.floor(X)  # Округление вниз
X_ceil = np.ceil(X)    # Округление вверх
X_trunc = np.trunc(X)  # Усечение дробной части
X_round = np.round(X)  # Округление до ближайшего целого

# 4. Вычисляем матрицу размера n на 3 с суммами, средними значениями и среднеквадратичными отклонениями
sums = np.sum(X, axis=1)
means = np.mean(X, axis=1)
std_devs = np.std(X, axis=1)
result_matrix = np.vstack((sums, means, std_devs)).T

# Выводим результаты
print("Матрица X:\n", X)
print("Матрица с замененными отрицательными значениями:\n", X_non_negative)
print("Матрица Y:\n", Y)
print("Целочисленная часть (floor):\n", X_floor)
print("Целочисленная часть (ceil):\n", X_ceil)
print("Целочисленная часть (trunc):\n", X_trunc)
print("Целочисленная часть (round):\n", X_round)
print("Матрица с суммами, средними значениями и среднеквадратичными отклонениями:\n", result_matrix)
```

**Задание 4.** Вычислите так называемую случайную матрицу Вигнера $W$ размера $n \times n$. Для этого вычислите случайную матрицу $X$ из нормально распределенных чисел (размера $n$ на $n$, $\mu=0$, $\sigma=1$). Матрица $W$ определяется формулой: $$ \frac {(X + X^T)}{\sqrt{2n}}.$$ Вычислите среднее значение и среднеквадратичное отклонение элементов полученной матрицы $W$. 

```python
#Этот код создает симметричную матрицу Вигнера размера n x n, используя случайные нормально распределенные числа. Затем вычисляются среднее 
#значение и стандартное отклонение элементов этой матрицы.
# ваш код здесь
import numpy as np

def wigner_matrix(n):
    # Создаем случайную матрицу X из нормально распределенных чисел
    X = np.random.normal(0, 1, (n, n))
    
    # Вычисляем матрицу W по формуле
    W = (X + X.T) / np.sqrt(2 * n)
    
    # Вычисляем среднее значение и среднеквадратичное отклонение элементов матрицы W
    mean_W = np.mean(W)
    std_W = np.std(W)
    
    return W, mean_W, std_W

# Пример использования функции
n = 5  # Размер матрицы
W, mean_W, std_W = wigner_matrix(n)

print("Матрица W:\n", W)
print("Среднее значение элементов матрицы W:", mean_W)
print("Среднеквадратичное отклонение элементов матрицы W:", std_W)
```

**Задание 5.** Повторите 1000 раз следующий эксперимент: сгенерируйте две матрицы размера $n \times n$ из стандартного нормального распределения, перемножьте их (обычным матричным умножением) и найдите максимальный элемент. Какое среднее значение по экспериментам у максимальных элементов? 25- и 75-процентные квантили? Постройте в виде гистограммы распределение максимальных элементов. Для вычисления квантили воспользуйтесь функцией `np.percentile`

```python
#Этот код проводит 1000 экспериментов, где генерируются две случайные матрицы, перемножаются, и для каждой новой матрицы находит 
# максимальный элемент.Затем выводятся среднее значение этих максимальных элементов, 25- и 75-процентные квантили, 
# и строится гистограмма распределения этих максимальных значений.
# ваш код здесь
import numpy as np
import matplotlib.pyplot as plt

n = 10

def experiment(n, num_experiments=1000):
    max_elements = []
    
    for _ in range(num_experiments):
        # Генерация двух случайных матриц размера n x n
        A = np.random.normal(0, 1, (n, n))
        B = np.random.normal(0, 1, (n, n))
        
        # Перемножение матриц
        C = np.dot(A, B)
        
        # Нахождение максимального элемента в матрице C
        max_element = np.max(C)
        max_elements.append(max_element)
    
    # Вычисление среднего значения максимальных элементов
    mean_max = np.mean(max_elements)
    
    # Вычисление 25- и 75-процентных квантилей
    q25 = np.percentile(max_elements, 25)
    q75 = np.percentile(max_elements, 75)
    
    # Построение гистограммы распределения максимальных элементов
    plt.hist(max_elements, bins=30, edgecolor='black')
    plt.title('Распределение максимальных элементов')
    plt.xlabel('Максимальный элемент')
    plt.ylabel('Частота')
    plt.show()
    
    return mean_max, q25, q75

# Пример использования функции
n = 5  # Размер матрицы
mean_max, q25, q75 = experiment(n)

print("Среднее значение максимальных элементов:", mean_max)
print("25-процентный квантиль:", q25)
print("75-процентный квантиль:", q75)
```

**Задание 6.** Сгенерируйте $n$ чисел из распределения Пуассона с параметром $\lambda$. Сформируйте массив, в котором в i-м элементе будет записано количество сгенерированных чисел, равных i. Постройте гистограмму эмпирического распределения. Реализуйте функцию, которая принимает на вход numpy-массив целых чисел a (подобный полученному), и генерирует массив, в котором число i встречается a[i] раз.

**Подсказка:** обратите внимание на функции `np.bincount` и `np.repeat`

```python
#Этот код:#Генерирует n случайных чисел из распределения Пуассона с параметром λ.
#Строит гистограмму частот этих чисел.
#Создает массив, в котором каждое число встречается столько раз, сколько показала гистограмма.
#Таким образом, код сначала исследует распределение Пуассона, а затем восстанавливает массив на основе полученных частот.
# ваш код здесь
import numpy as np
import matplotlib.pyplot as plt

def poisson_experiment(n, lam):
    # Генерация n чисел из распределения Пуассона с параметром λ
    poisson_samples = np.random.poisson(lam, n)
    
    # Формирование массива, где i-й элемент - количество чисел, равных i
    counts = np.bincount(poisson_samples)
    
    # Построение гистограммы эмпирического распределения
    plt.bar(range(len(counts)), counts, align='center', alpha=0.7)
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.title('Гистограмма эмпирического распределения Пуассона')
    plt.show()
    
    return counts

def generate_array_from_counts(counts):
    # Генерация массива, где число i встречается counts[i] раз
    result_array = np.repeat(np.arange(len(counts)), counts)
    return result_array 

# Пример использования функций
n = 1000  # Количество чисел
lam = 5   # Параметр λ распределения Пуассона

counts = poisson_experiment(n, lam)
result_array = generate_array_from_counts(counts)

print("Массив counts:", counts)
print("Сгенерированный массив:", result_array)
```

Загрузите и перемасштабируйте таблицу с векторами признаков из встроенного набора `"Wine recognition dataset"` [https://scikit-learn.org/stable/datasets/toy_dataset.html#wine-dataset ]

```python
#Этот код загружает набор данных "Wine", вычисляет среднее и стандартное отклонение для каждого признака, и затем масштабирует данные так, 
#чтобы каждый столбец имел нулевое среднее и единичное стандартное отклонение.
# ваш код здесь
import numpy as np
from sklearn.datasets import load_wine 

def scale(X):
    # Вычисляем среднее значение и стандартное отклонение для каждого столбца
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0) + 1e-6  # Добавляем малую константу к стандартному отклонению
    
    # Масштабируем каждый столбец
    X_scaled = (X - mean) / std
    return X_scaled

# Загрузка набора данных Wine
data = load_wine()
X = data.data

# Применение функции масштабирования
X_scaled = scale(X)

print("Масштабированные данные:\n", X_scaled)
```

**Задание 9.** Реализуйте функцию, вычисляющую коэффициенты линейной регрессии $a_1$ и $a_0$ для зависимости вида $y_i(x)=a_1 x_i + a_0$. Для этого обратитесь к материалу второй лекции курса, где был разобран подобный пример. На вход функция должна принимать два вектора $x$ и $y$. Результат работы должен представлять два значения коэффициентов $a_1$ и $a_0$.

```python
#Этот код вычисляет коэффициенты линейной регрессии для данных x и y, используя формулы для наклона a_1 и пересечения с осью a_0. 
#Эти коэффициенты определяют уравнение прямой вида 𝑦=𝑎1𝑥+𝑎0 y=a1 x+a 0 которая наилучшим образом описывает зависимость между x и y.
# ваш код здесь   
def linear_regression_coefficients(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x_i * y_i for x_i, y_i in zip(x, y))
    sum_x_squared = sum(x_i ** 2 for x_i in x)
    
    a_1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    a_0 = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x ** 2)
    
    return a_1, a_0

# Пример использования функции
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 6, 8]
a_1, a_0 = linear_regression_coefficients(x, y)
print(f"Коэффициенты линейной регрессии: a_1 = {a_1}, a_0 = {a_0}")
```

В качестве $x$ используйте значения признака индекса массы тела (`bmi`), в качестве $y$ - значения по ключу `target`. Вычислите коэффициенты $a_1$ и $a_0$. Постройте график с осью абсцисс, отвечающей за значения `bmi`, осью ординат со значениями `target`. В виде 442 точек нанесите примеры из загруженных данных, в виде прямой представьте предсказания модели. Подпишите оси координат и метки значений на осях

```python
#Этот код загружает набор данных о диабете, извлекает признак "bmi" и целевую переменную "target". Затем вычисляет коэффициенты линейной регрессии
#для связи между "bmi" и прогрессированием диабета. Наконец, строит график, отображая точки данных и линию линейной регрессии, которая показывает, 
#как "bmi" предсказывает прогрессирование диабета.
# ваш код здесь  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# Загрузка набора данных
diabetes = datasets.load_diabetes()
data = pd.DataFrame(data=np.c_[diabetes['data'], diabetes['target']], columns=diabetes['feature_names'] + ['target'])

# Извлечение признака 'bmi' и целевой переменной 'target'
x = data['bmi'].values
y = data['target'].values
```

```python
def linear_regression_coefficients(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x_i * y_i for x_i, y_i in zip(x, y))
    sum_x_squared = sum(x_i ** 2 for x_i in x)
    
    a_1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    a_0 = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x ** 2)
    
    return a_1, a_0

# Вычисление коэффициентов
a_1, a_0 = linear_regression_coefficients(x, y)
print(f"Коэффициенты линейной регрессии: a_1 = {a_1}, a_0 = {a_0}")
```

```python
# Построение графика
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Данные')
plt.plot(x, a_1 * x + a_0, color='red', label='Линейная регрессия')
plt.xlabel('BMI')
plt.ylabel('Прогрессирование диабета')
plt.title('Линейная регрессия для предсказания прогрессирования диабета')
plt.legend()
plt.show()
```
